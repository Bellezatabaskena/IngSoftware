```python
print("* * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * ")
print("* * * * * * * * *   C A L C U L A D O R A  D E  G A S T O S   * * * * * * * * * * * * * ")
print("* * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * " + "\n")

# Aqui se le pide al usuario ingresar su nombre
usuario  = input("Hola,cual es tu nombre?: " + "\n")
print("Bienvenido a la Calculadora de Gastos, " + (usuario) + "\n");
print("* * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * " + "\n")

# En este arreglo se guardaran las cantidades para cada dia de la semana
numeros = []

# Con estas variables se tomaran en cuenta las cantidades mas grande y mas pequeña
cant_mayor = None
cant_menor = None
dia_mayor = None
dia_menor = None

# Se definen los dias de la semana
dias_semana = ["Lunes", "Martes", "Miercoles", "Jueves", "Viernes", "Sábado", "Domingo"]

# Se le pide al usuario que ingrese una cantidad por dia de la semana
for day in dias_semana:
    while True:
        try:
            # Aqui se pide ingresar una cantidad por dia de la semana y se redondea a 2 decimales despues del punto, unicamente
            num = round(float(input(f"Por favor ingrese la cantidad total del dia {day}: ")), 2)
            numeros.append(num)
            print("* * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * " + "\n")

            # Mediante la condicional if, se toman en cuenta las cantidades menor y mayor.
            if cant_mayor is None or num > cant_mayor:
                cant_mayor = num
                dia_mayor = day

            if cant_menor is None or num < cant_menor:
                cant_menor = num
                dia_menor = day

             #se rompera el ciclo si se ingresa algun otro caracter que no sea numero
            break
        except ValueError:
            print("Esa entrada es invalida, por favor ingrese una cantidad entera o decimal.")
            print("* * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * ")

# Se calcula el total de las cantidades ingresadas
total = round(sum(numeros), 2)  # Se redondea a 2 decimales despues del punto.

# Se ordenan las cantidades de manera ascendente
sorted_numbers = sorted(numeros)

# Se ordenan las cantidades de manera descendente
reverse_sorted_numbers = sorted(numeros, reverse=True)

# Se imprimen las listas ordenadas, suma, y los dias con mayor y menor cantidad.
print("* * * * * * * * * * * * * * * *  R E S U L T A D O S  * * * * * * * * * * * * * * * * * " + "\n")
print("Cantidades ordenadas de forma ascendente:" + "\n", sorted_numbers)
print("* * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * ")
print("Cantidades ordenadas de forma descendente:" + "\n", reverse_sorted_numbers)
print("* * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * ")
print("Suma total de gastos:", total)
print("* * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * ")
print(f"Dia de la semana con el gasto mayor: {dia_mayor} ({cant_mayor})")
print("* * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * ")
print(f"Dia de la semana con el gasto menor: {dia_menor} ({cant_menor})")
print("* * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * " + "\n")
print("* * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * ")
print("* * * * * * * * * * * * *  F I N  D E L  P R O G R A M A  * * * * * * * * * * * * * * * ")
print("* * * * * * * * * * * * * * * *   G R A C I A S   * * * * * * * * * * * * * * * * * * * ")
print("* * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * * ")
```
